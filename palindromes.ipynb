```python
def is_palindrome(n):
    if str(n) == str(n)[::-1]:
        return True;
    else:
        return False

palindrome = [[i, j, i*j] for i in range(1,10000) for j in range(i,10000) if is_palindrome(i*j)]
```

```python
xs = [p[0] for p in palindrome]
ys = [p[1] for p in palindrome]
length = [len(str(p[2])) for p in palindrome]
```

```python
import numpy as np
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(20,20))
    
x = [p[0] for p in palindrome]
y = [p[1] for p in palindrome]
length = [len(str(p[2])) for p in palindrome]

plt.scatter(x, y, c=length)
plt.scatter(y, x, c=length)

plt.xticks(fontsize = 20)
plt.yticks(fontsize = 20)
plt.xlim(0,10000)
plt.ylim(0,10000)
plt.show()
```
